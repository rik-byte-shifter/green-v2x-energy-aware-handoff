```python
import json
import matplotlib.pyplot as plt
import numpy as np

with open('results/parameter_sweep.json') as f:
    data = json.load(f)

x = [r['availability_pct'] for r in data]
y = [r['energy_saving_pct'] for r in data]
colors = [r['coverage_gap_pct'] for r in data]

plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, c=colors, cmap='RdYlGn_r', s=100, edgecolors='k')
plt.colorbar(scatter, label='Coverage Gap (%)')
plt.xlabel('Service Availability (%)')
plt.ylabel('Energy Saving vs RSSI (%)')
plt.title('Energy-Aware Handoff: Pareto Frontier')
plt.grid(True, alpha=0.3)
plt.axhline(20, color='gray', linestyle='--', label='Min Publishable Energy Saving')
plt.axvline(85, color='gray', linestyle='--', label='Min Publishable Availability')
plt.legend()
plt.tight_layout()
plt.savefig('results/pareto_frontier.png', dpi=300)
plt.show()
```
